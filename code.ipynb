```python
import yaml
from ultralytics import YOLO

# Define paths
train_path = r"/content/drive/MyDrive/Dataset/train"
val_path = r"/content/drive/MyDrive/Dataset/valid"
data_yaml_path = r"/content/drive/MyDrive/Dataset/data.yaml"

# Load and verify data.yaml
with open(data_yaml_path, 'r') as file:
    data = yaml.safe_load(file)

# Load or initialize YOLOv8 model
model = YOLO('yolov8m.pt')  # Using medium model for better generalization

# Train the model with improved settings
print("Starting model training...")
model.train(
    data=data_yaml_path,
    imgsz=640,
    epochs=50,  # Increased epochs for better learning
    batch=16,
    name='signature_detection_train_improved',
    augment=True,  # Enable data augmentation
    lr0=0.01,  # Learning rate
    weight_decay=0.0005,  # Regularization to prevent overfitting
    val=True  # Validate during training
)
print("Model training completed.")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import os

# Load the trained model
model = YOLO('runs/detect/signature_detection_train_improved/weights/best.pt')

# Function to extract all signature patterns and save annotated image
def extract_signature_patterns(image_path, model, conf_threshold=0.2):
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not load image at {image_path}")
        return None, None

    # Predict and save annotated image
    results = model.predict(image_path, conf=conf_threshold, save=True, save_txt=False, save_conf=True)
    output_dir = 'runs/detect/predict'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    annotated_path = os.path.join(output_dir, os.path.basename(image_path))

    patterns = []
    for result in results:
        boxes = result.boxes
        if len(boxes) > 0:
            for box in boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                signature_crop = img[y1:y2, x1:x2]
                # Resize and convert to grayscale
                signature_crop = cv2.resize(signature_crop, (224, 224))
                signature_gray = cv2.cvtColor(signature_crop, cv2.COLOR_BGR2GRAY)
                # Edge detection for pattern analysis
                edges = cv2.Canny(signature_gray, 100, 200)
                # Find contours
                contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                if contours:
                    patterns.append((signature_crop, contours))
    if not patterns:
        print(f"No signature patterns detected in {image_path}")
    return patterns, annotated_path

# Function to compare signature patterns
def compare_signature_patterns(patterns1, patterns2):
    if not patterns1 or not patterns2:
        print("Comparison failed: One or both images do not contain detectable signature patterns.")
        return False

    # Compare each pattern from image1 with each pattern from image2
    for crop1, contours1 in patterns1:
        for crop2, contours2 in patterns2:
            # Calculate contour similarity using matchShapes
            if contours1 and contours2:
                similarity = cv2.matchShapes(contours1[0], contours2[0], cv2.CONTOURS_MATCH_I3, 0.0)
                print(f"Pattern similarity score: {similarity:.4f}")
                # Threshold for considering patterns similar (lower value = more similar)
                if similarity < 0.1:  # Adjust threshold as needed
                    return True
    return False

# Main function to run the comparison
def main():
    # Get image paths from user
    image1_path = input("Enter the path to the first image: ")
    image2_path = input("Enter the path to the second image: ")

    if not os.path.exists(image1_path):
        print(f"Error: Image at {image1_path} does not exist.")
        return
    if not os.path.exists(image2_path):
        print(f"Error: Image at {image2_path} does not exist.")
        return

    # Extract patterns from both images
    patterns1, _ = extract_signature_patterns(image1_path, model)
    patterns2, _ = extract_signature_patterns(image2_path, model)

    # Compare the signature patterns
    are_similar = compare_signature_patterns(patterns1, patterns2)
    if are_similar:
        print("The signatures in the two images are same (similar patterns detected).")
    else:
        print("The signatures in the two images are not same (no similar patterns detected).")

if __name__ == "__main__":
    main()
```
